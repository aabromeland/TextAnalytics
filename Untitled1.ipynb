```python
import requests  #The library to download html file from internet
from lxml import html      #XML/HTML parser
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np


URL = 'https://www.dailywire.com/read'
page = requests.get(URL) #Download the HTML file through URL.
root = html.fromstring(page.content) #Parse the HTML source code and create the tree object.


links = [s.attrib['href'] for s in root.xpath('//div/article/a')]
headlines = [s.text_content() for s in root.xpath('//div/article/a/div[2]/h3')]
authors = [s.text_content() for s in root.xpath('//div/article/a/div[2]/div')]
links = links[1:]
authors = authors[1:]

print(len(links))
print(len(headlines))
print(len(authors))
df = pd.DataFrame({'links': links, 'headlines': headlines,'authors': authors})
df
```

```python
from afinn import Afinn
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt  

afinn = Afinn(emoticons=True)
df["AFINN"]=[afinn.score(s) for s in df.headlines]

# TextBlob Sentiment Analysis Scores
df["TextBlob"]=[TextBlob(s).sentiment.polarity for s in df.headlines]

# VADER Sentiment Analysis Scores
analyzer=SentimentIntensityAnalyzer()
df["VADER"]=[analyzer.polarity_scores(s)['compound'] for s in df.headlines]

df['AFINN_SCALE'] = MinMaxScaler().fit_transform(np.array(df['AFINN']).reshape(-1,1))

%matplotlib inline
plt.figure(figsize=(15, 6))

plt.plot(df.index, df.AFINN_SCALE, "-r", label="AFINN Scaled")
plt.plot(df.index, df.TextBlob, "-b", label="TextBlob")
plt.plot(df.index, df.VADER, "-g", label="VADER")



plt.legend(loc="upper left")
plt.title("Comparison on Sentiment Score")
plt.xlabel("Date")
plt.ylabel("Sentiment Score (Moving Average with Window Size 50)")
plt.grid(axis='both')
plt.show()
```
