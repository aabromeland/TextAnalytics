### Remove Small Character Counts and Re-Run Summaries

```python
dedfClean = dedf[dedf['characters']>500].copy()
dedfClean.sort_values(by='filingDate',ascending=False,inplace=True)
dedfClean.reset_index(inplace=True,drop=True)
print(dedfClean.shape)
dedfClean
```

```python
print(f"Average character length: {dedfClean['characters'].mean()}")
print(f"Minimum character length: {dedfClean['characters'].min()}")
print(f"Maximum character length: {dedfClean['characters'].max()}")
```

## Creation of DTM's

### Unigrams

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
import pysentiment2 as ps

lm = ps.LM()
vectorizer = CountVectorizer(tokenizer = lm.tokenize)
DTM = vectorizer.fit_transform(dedfClean['text'])
```

```python
dffreq = pd.DataFrame({'Term': vectorizer.get_feature_names_out(),
                   'Frequency': DTM.sum(axis=0).tolist()[0]
                  })

dffreq.sort_values(by="Frequency",inplace=True,ascending=False)
dffreq.reset_index(inplace=True,drop=True)
dffreq.head(10)
```

```python
nltk_stopwords = nltk.corpus.stopwords.words("english")
vectorizer = CountVectorizer(tokenizer = lm.tokenize,stop_words=nltk_stopwords)
DTM = vectorizer.fit_transform(dedfClean['text'])
dffreq = pd.DataFrame({'Term': vectorizer.get_feature_names_out(),
                   'Frequency': DTM.sum(axis=0).tolist()[0]
                  })

dffreq.sort_values(by="Frequency",inplace=True,ascending=False)
dffreq.reset_index(inplace=True,drop=True)
dffreq.head(10)
```

### Bi-grams

```python
vectorizer = CountVectorizer(tokenizer = lm.tokenize,ngram_range=(2,2))
DTM = vectorizer.fit_transform(dedfClean['text'])
dffreq = pd.DataFrame({'Term': vectorizer.get_feature_names_out(),
                   'Frequency': DTM.sum(axis=0).tolist()[0]
                  })

dffreq.sort_values(by="Frequency",inplace=True,ascending=False)
dffreq.reset_index(inplace=True,drop=True)
dffreq.head(10)
```

```python
nltk_stopwords = nltk.corpus.stopwords.words("english")
vectorizer = CountVectorizer(tokenizer = lm.tokenize,stop_words=nltk_stopwords,ngram_range=(2,2))
DTM = vectorizer.fit_transform(dedfClean['text'])
dffreq = pd.DataFrame({'Term': vectorizer.get_feature_names_out(),
                   'Frequency': DTM.sum(axis=0).tolist()[0]
                  })

dffreq.sort_values(by="Frequency",inplace=True,ascending=False)
dffreq.reset_index(inplace=True,drop=True)
dffreq.head(10)
```
